```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('energydata_complete.csv')
df = df.drop(columns=['date', 'lights'])
```

```python
df.head()
```

```python
#Normalize the dataset to common scale using min-max scaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalized_df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
normalized_df.head()
```

```python
target = normalized_df['Appliances']
features = normalized_df.drop(columns=['Appliances'])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

linear_model = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features,target, test_size=0.3, 
                                                   random_state=42)
```

```python
#Question 12
linear_model.fit(X_train[['T2']], X_train.T6)
pred = linear_model.predict(X_test[['T2']])
#RSQUARED: R-Sqaured Score
rsquared = r2_score(y_test, pred)
print("R squared score: ", round(rsquared, 2))
```

```python
linear_model.fit(X_train, y_train)
```

```python
predicted_values = linear_model.predict(X_test)
```

```python
#Question 13
#MAE: Mean Absolute Error
mae = mean_absolute_error(y_test, predicted_values)
print("Mean Absolute Error (MAE): ", round(mae, 2))
```

```python
#Question 14
#RSS: Residual Sum of Squares
rss = np.sum(np.square(y_test - predicted_values))
print("Residual Sum of Squares (RSS): ", round(rss, 2))
```

```python
#Question 15
#RMSE: Root Mean Squared Error
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
print("Root Mean Squared Error: ", round(rmse, 3))
```

```python
#Question 16
#RSQUARED: R-Sqaured Score
rsquared = r2_score(y_test, predicted_values)
print("R squared score: ", round(rsquared, 2))
```

```python
#Question 17
def get_weights_df(model, feat, col_name):
    weights =  pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df
linear_weights_df = get_weights_df(linear_model, X_train, 'Linear_Model_Weight')
linear_weights_df
```

```python
#Question 18

#Ridge Regularization
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=.4)
ridge_reg.fit(X_train, y_train)
ridge_pred = ridge_reg.predict(X_test)
#Root Mean Square Error
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, ridge_pred))
round(rmse, 3)
```

```python
#Question 19

#LASSO Regularization
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(X_train, y_train)
lasso_weights_df = get_weights_df(lasso_reg, X_train, 'Lasso_Model_Weight')
lasso_weights_df
```

```python
#Question 20

#RMSE with the Lasso Regression
lasso_pred = lasso_reg.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, lasso_pred))
round(rmse, 3)
```
